## Preprocessing of Data
- Train | Test Split, Scalling

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Exited', axis = 1)
y = df['Exited']
```

```python
seed = 42
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=seed)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

## Modelling & Model Performance

### Import related libraries

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from scikeras.wrappers import KerasClassifier, KerasRegressor
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold 
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_roc_curve, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.optimizers import Adam
```

### Creating Model

### without class_weigth

```python
""" class_weight: Optional dictionary mapping class indices (integers)
        to a weight (float) value, used for weighting the loss function
        (during training only).
        This can be useful to tell the model to
        "pay more attention" to samples from
        an under-represented class."""

# define sequential model

tf.random.set_seed(seed) 

model = Sequential()
model.add(Dense(15, input_shape=(10,), activation='relu'))
model.add(Dense(10,  activation='relu'))
model.add(Dense(5, activation = 'relu'))
model.add(Dense(1, activation='sigmoid')) # we use sigmoid for binary output
model.summary()


# compile the model

opt = Adam(lr = 0.003)
model.compile(optimizer = opt, loss = 'binary_crossentropy',metrics = ['accuracy'])

early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 20)

# now we fit our model to training data

history = model.fit(X_train, y_train, epochs=250, validation_split = 0.2, batch_size=64, callbacks=[early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

#### Evaluate

```python
model.evaluate(X_test,y_test)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

#print classification_report

y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))

# ploting the confusion metrix plot

conf_mat = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (17,7))
sns.heatmap(conf_mat, annot=True,fmt='d')
plt.xlabel('Predicted_number')
plt.ylabel('True_number')
```

### with class_weigth

Investigate how the "class_weight" hyper-parameter is used in a Neural Network.

```python
from sklearn.utils import class_weight
```

```python
class_weights = class_weight.compute_class_weight('balanced',
                                                 classes= np.unique(y_train),
                                                 y = y_train)
```

```python
class_weights
```

```python
""" class_weight: Optional dictionary mapping class indices (integers)
        to a weight (float) value, used for weighting the loss function
        (during training only).
        This can be useful to tell the model to
        "pay more attention" to samples from
        an under-represented class."""


# define sequential model

model_weighted = Sequential()
model_weighted.add(Dense(15, input_shape=(10,), activation='relu'))
model_weighted.add(Dense(10, activation='relu'))
model_weighted.add(Dense(5,  activation = 'relu'))
model_weighted.add(Dense(1, activation='sigmoid')) # we use sigmoid for binary output
model_weighted.summary()


# compile the model

opt = Adam(lr = 0.005)
model_weighted.compile(optimizer = opt, loss = 'binary_crossentropy',metrics = ['accuracy'])

early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 15)

# now we fit our model to training data

class_weight = {0: 0.62791532, 1:2.45441795}

history = model_weighted.fit(X_train, y_train, epochs=250, 
                    validation_split = 0.2,
                    batch_size=64,  
                    callbacks = [early_stop],
                    class_weight=class_weight)
                    

```

```python
loss_df = pd.DataFrame(model_weighted.history.history)
loss_df.plot()
```

#### Evaluate

```python
model_weighted.evaluate(X_test,y_test)
```

```python
loss, accuracy = model_weighted.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

#print classification_report
y_pred = (model_weighted.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))

# ploting the confusion metrix plot

conf_mat = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (17,7))
sns.heatmap(conf_mat, annot=True,fmt='d')
plt.xlabel('Predicted_number')
plt.ylabel('True_number')
```

## GridSearchCV

```python
from tensorflow.keras.optimizers import Adam, Adadelta, RMSprop, Nadam
```

```python
def build_classifier(optimizer, learn_rate):
    classifier = Sequential()
    classifier.add(Dense(units = 15, activation = 'relu'))
    classifier.add(Dense(units = 10, activation = 'relu'))
    classifier.add(Dense(units = 10, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer(learn_rate), loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
early_stop = EarlyStopping(monitor = "val_accuracy", mode = "auto", verbose = 1, patience = 15)
```

```python
tf.random.set_seed(seed)

classifier_model = KerasClassifier(build_fn = build_classifier, validation_split = 0.1, epochs = 200)

parameters = {'batch_size': [32, 64],
              'optimizer': [Adam, RMSprop, Adadelta, Nadam],
              'learn_rate': [0.001, 0.003, 0.005]}

grid_model = GridSearchCV(estimator = classifier_model,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10,
                          n_jobs = -1,
                          verbose = 1)

grid_model.fit(X_train, y_train, callbacks = [early_stop])
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
grid_result = pd.DataFrame(grid_model.cv_results_)[["mean_test_score", "std_test_score", "rank_test_score", "params"]]
grid_result
```

```python

#print classification_report
y_pred = (grid_model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))

# ploting the confusion metrix plot

conf_mat = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (17,7))
sns.heatmap(conf_mat, annot=True,fmt='d')
plt.xlabel('Predicted_number')
plt.ylabel('True_number')
```

## Final Model and Model Deployment

```python
import pickle
pickle.dump(scaler, open("scaler_churn", 'wb'))
```

```python
tf.random.set_seed(seed)

# define sequential model

final_model = Sequential()
final_model.add(Dense(15, input_shape=(10,), activation='relu'))
final_model.add(Dense(10, activation='relu'))
final_model.add(Dense(5,  activation = 'relu'))
final_model.add(Dense(1, activation='sigmoid')) # we use sigmoid for binary output
final_model.summary()


# compile the model

opt = Adam(lr = 0.005)
final_model.compile(optimizer = opt, loss = 'binary_crossentropy',metrics = ['accuracy'])

# now we fit our model to training data

class_weight = {0: 0.62791532, 1:2.45441795}
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 15)
```

```python
final_model.fit(x = X_train, y = y_train, validation_data = (X_test, y_test), batch_size =64, epochs = 250, verbose = 1,
          callbacks = [early_stop],class_weight=class_weight) 
```

```python
loss_df = pd.DataFrame(final_model.history.history)
loss_df.plot()
```

```python
loss, accuracy = final_model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

#print classification_report
y_pred = (final_model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))

# ploting the confusion metrix plot

conf_mat = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize = (17,7))
sns.heatmap(conf_mat, annot=True,fmt='d')
plt.xlabel('Predicted_number')
plt.ylabel('True_number')
```

```python
final_model.save('model_churn.h5')
```
